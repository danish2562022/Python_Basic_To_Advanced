```python
from matplotlib import pyplot as plt
plt.style.use("fivethirtyeight")
```

```python
plt.title("PieChart")
slices=[60,40]
plt.pie(slices)
```

```python
plt.title("PieChart")
labels=["Sixty","Forty"]
slices=[60,40]
plt.pie(slices,labels=labels)
```

```python
plt.title("PieChart")
labels=["Sixty","Forty"]
slices=[120,80]
plt.pie(slices,labels=labels)
```

```python
plt.title("PieChart")
labels=["Sixty","Forty"]
slices=[120,80]
plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]

plt.pie(slices,labels=labels,wedgeprops={'edgecolor':'black'})
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]
explode=[0,0,0,0.2]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'})
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]
explode=[0,0,0,1]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'})
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]
explode=[0,0,0,0.2]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True)
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]
explode=[0,0,0,0.2]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True,startangle=90)
```

```python
plt.title("PieChart")
labels=["one hundred and twenty","Eighty","Thirty","Twenty"]
slices=[120,80,30,20]
explode=[0,0,0,0.2]

plt.pie(slices,labels=labels,explode=explode,wedgeprops={'edgecolor':'black'},shadow=True,startangle=90,autopct="%1.1f%%")
```
