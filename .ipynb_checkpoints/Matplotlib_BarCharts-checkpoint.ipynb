```python
import pandas as pd
data=pd.read_csv("https://raw.githubusercontent.com/CoreyMSchafer/code_snippets/master/Python/Matplotlib/02-BarCharts/data.csv")
```

```python
data.head()
```

```python
from collections import Counter
c=Counter(["Python","javascript"])
```

```python
c
```

```python
c.update(["C++","Python"])
c
```

```python
language_counter=Counter()
for row in data['LanguagesWorkedWith']:
    
    language_counter.update(row.split(';'))
```

```python
print(language_counter.most_common(15))
```

```python
languages=[]
popularity=[]
for item in language_counter.most_common(15):
    languages.append(item[0])
    popularity.append(item[1])
print(languages)
print(popularity)
```

```python
plt.bar(languages,popularity)
plt.title("Most popular Language")
plt.xlabel("Programming Language")
plt.ylabel("Popularity")
plt.tight_layout()
plt.show()
```

```python
plt.barh(languages,popularity)
plt.title("Most popular Language")
plt.ylabel("Programming Language")
plt.xlabel("Popularity")
plt.tight_layout()
plt.grid()
plt.show()
```

```python
languages.reverse()
popularity.reverse()
plt.barh(languages,popularity)
plt.title("Most popular Language")
plt.ylabel("Programming Language")
plt.xlabel("Popularity")
plt.tight_layout()
plt.grid()
plt.show()
```
