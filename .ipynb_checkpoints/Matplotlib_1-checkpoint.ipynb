```python
from matplotlib import pyplot as plt
```

```python
x=[23,24,25,26]
y=[1000,2000,3000,1900]
```

```python
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.show()
```

```python
y2=[500,1500,2500,3500]
```

```python
plt.plot(x,y)
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2)
plt.legend(["y","y1"])
plt.show()
```

```python
plt.plot(x,y,label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,label="y1")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,'k--',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,label="y1")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='k',linestyle='--',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='b',label="y1")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='k',linestyle='--',marker='.',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='b',marker='o',label="y1")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='#444444',linestyle='--',marker='.',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='#5a7d9a',marker='o',label="y1")
plt.legend()
plt.show()
```

```python
plt.plot(x,y,color='#444444',linestyle='--',marker='.',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='#5a7d9a',marker='o',label="y1")
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
print(plt.style.available)
plt.plot(x,y,color='#444444',linestyle='--',marker='.',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='#5a7d9a',marker='o',label="y1")
plt.legend()
plt.tight_layout()
plt.grid()
plt.show()
```

```python
plt.style.use('ggplot',)
plt.plot(x,y,color='#444444',linestyle='--',marker='.',label="y")
plt.xlabel('ages')
plt.ylabel('salary')
plt.title("Salary vs age")
plt.plot(x,y2,color='#5a7d9a',marker='o',label="y1")
plt.legend()
plt.show()
```
